```python
#detect outliers with Z-score
outliers = []
def detect_outliers(data):
    threshold =3
    mean = np.mean(data)
    std = np.std(data)
    
    for i in data:
        z_score=(i-mean)/std
        if np.abs(z_score)>threshold:
            outliers.append(i)
    return outliers 

```

```python
detect_outliers(df['season17_18'])
```
